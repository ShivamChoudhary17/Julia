```julia
df_x = convert(Array, df_new[:, [1,2,3,4]]);
```

```julia
df_y = convert(Array, df_new[:, [5]]);
```

```julia
using ScikitLearn
```

```julia
@sk_import linear_model : LogisticRegression
```

```julia
log_reg_model = LogisticRegression()
```

```julia
fit!(log_reg_model, df_x, df_y)
```

```julia
prediction = predict(log_reg_model, df_x);
```

```julia
plot(prediction)
```

```julia
@sk_import metrics : accuracy_score
```

```julia
accuracy = accuracy_score(prediction, df_y)
```

```julia
@sk_import tree : DecisionTreeClassifier
```

```julia
tree_model = DecisionTreeClassifier()
```

```julia
fit!(tree_model, df_x, df_y)
```

```julia
tree_prediction = predict(tree_model, df_x);
```

```julia
tree_accur = accuracy_score(tree_prediction, df_y)
```

```julia
plot(tree_prediction)
```
